```python
row=[]
for i in range (len(mf_df)):
    for j in range (len(hpv_df)):
        match=mf_df['filename'][i].split('.')[0].split('-')[0:3]
        match= match[0]+'-'+match[1]+'-'+match[2]
        match1=mf_df['filename'][i].split('.')[0].split('-')[-1]
#         print(match1[1])
        if match == hpv_df['CaseID'][j] and match1[1]== 'X':
            print(f'match is {i+2} and hpv is {j+2}')
            row.append(i)
            
        
```

```python
df = mf_df.iloc[row[0]:row[0]+1,:]

for k in range (1, len(row)):
    
    df_2=mf_df.iloc[row[k]:row[k]+1,:]
    df=pd.concat([df, df_2], axis=0)
```
